```python
# ייבוא ספריות
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# טען את הקובץ
df = pd.read_csv("Full_Data.csv")

# חישוב TotalVotes (סכום הצבעות חיוביות ושליליות)
df['TotalVotes'] = df['Upvotes'] + df['Downvotes']

# טיפול בערכים חסרים
df['TotalVotes'] = df['TotalVotes'].fillna(0)

# יצירת EngagementRatio לפני מחיקת העמודות
df['EngagementRatio'] = df.apply(
    lambda row: row['Number of Comments'] / row['TotalVotes'] if row['TotalVotes'] > 0 else 0,
    axis=1
)


###############################
#סף מקור היה 0.3 להצבעות
#סף מקור לתגובות היה 0.4
###############################
# חישוב ספים על פי קוונטילים שנבחרו
vote_thr = df['TotalVotes'].quantile(0.5)   # קוונטיל 40%

com_thr = df['Number of Comments'].quantile(0.7)   # קוונטיל 40%

# יצירת עמודת buzz – רק אם שני התנאים מתקיימים
# df['buzz'] = ((df['TotalVotes'] >= vote_thr) & (df['Number of Comments'] >= com_thr)).astype(int)
df['buzz'] = ((df['TotalVotes'] >= vote_thr) | (df['Number of Comments'] >= com_thr)).astype(int)


# שמירת גרסה מלאה לניתוח גרפים (כולל כל העמודות)
df.to_csv("graf_buzz.csv", index=False)
print("✅ נוצר קובץ graf_buzz.csv עם כל עמודות העזר, buzz וכל היחסים, לניתוח גרפים.")


# מחיקת כל עמודות העזר ששימשו לחישוב
columns_to_drop = ['Upvotes', 'Downvotes']
df_cleaned = df.drop(columns=columns_to_drop)

# שמירת הקובץ החדש
df_cleaned.to_csv("data_with_buzz.csv", index=False)
print("✅ נוצר קובץ data_with_buzz.csv עם עמודת buzz מאוזנת ועמודות עזר הוסרו.")

# הדפסת סיכום קצר של ההתפלגות
print(df_cleaned['buzz'].value_counts(normalize=True).apply(lambda x: f"{x:.2%}"))
```
