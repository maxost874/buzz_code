# Bag of Words

זו אחת מהשיטות הכי בסיסיות ב־NLP לייצוג טקסט:

כל טקסט הופך לרשימה של ספירת מילים – בלי קשר לסדר שלהן.

```python
df_bow = df_model.copy()
```

```python
# המרת כל הדאטה ל-Bag of Words (CountVectorizer)
vectorizer = CountVectorizer(max_features=10000)
X_bow = vectorizer.fit_transform(df_bow['FullText'])
y = df_bow['buzz']
```

```python
# פיצול ל-Train/Test
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, y, test_size=0.3, random_state=42, stratify=y)
```

* Naive Bayes

```python
# יצירת ואימון המודל
nb = MultinomialNB()
nb.fit(X_train_bow, y_train)

# חיזוי
y_pred_nb = nb.predict(X_test_bow)

# דוחות
print("Accuracy:", accuracy_score(y_test, y_pred_nb))

print("=== Multinomial Naive Bayes על Bag of Words ===")
print(classification_report(y_test, y_pred_nb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_nb))
```

*  Random Forest

```python
# יצירת ואימון המודל
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train_bow, y_train)

# חיזוי
y_pred_rf = rf.predict(X_test_bow)

# דוחות
print("Accuracy:", accuracy_score(y_test, y_pred_rf))

print("=== Random Forest על Bag of Words ===")
print(classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

* XGBoost

```python
# יצירת ואימון המודל
xgb = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss',
    max_depth=5,
    n_estimators=100
)
xgb.fit(X_train_bow, y_train)

# חיזוי
y_pred_xgb = xgb.predict(X_test_bow)

# דוחות
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))

print("=== XGBoost על Bag of Words ===")
print(classification_report(y_test, y_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```

#  TF-IDF (Term Frequency – Inverse Document Frequency)


נותנת משקל גבוה למילים חשובות שמופיעות הרבה במסמך אבל מעט במסמכים אחרים.

היא מורכבת משני שלבים:
TF – כמה פעמים מילה מופיעה בטקסט (כמו BoW).

IDF – מה הופך אותה לנדירה. אם מילה מופיעה כמעט בכל הטקסטים, המשקל שלה ↓.

```python
df_tfidf = df_model.copy()
```

```python
# המרת כל הדאטה ל-TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=10000)
X_tfidf = tfidf_vectorizer.fit_transform(df_tfidf['FullText'])
y = df_tfidf['buzz']
```

```python
# פיצול ל-Train/Test
X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.35, random_state=42, stratify=y)
```

* Random Forest

```python
# יצירת ואימון המודל
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train_tfidf, y_train)

# חיזוי
y_pred_rf = rf.predict(X_test_tfidf)

# דוחות ביצועים
print("Accuracy:", accuracy_score(y_test, y_pred_rf))

print("=== Random Forest על TF-IDF ===")
print(classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

* XGBoost

```python

# יצירת ואימון המודל
xgb = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss',
    max_depth=5,
    n_estimators=100
)
xgb.fit(X_train_tfidf, y_train)

# חיזוי
y_pred_xgb = xgb.predict(X_test_tfidf)

# דוחות ביצועים
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))

print("=== XGBoost על TF-IDF ===")
print(classification_report(y_test, y_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```

* Naive Bayes

```python
# יצירת ואימון המודל
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)

# חיזוי
y_pred_nb = nb.predict(X_test_tfidf)

# דוחות ביצועים
print("Accuracy:", accuracy_score(y_test, y_pred_nb))

print("=== Multinomial Naive Bayes על TF-IDF ===")
print(classification_report(y_test, y_pred_nb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_nb))
```

# Word2Vec

Word2Vec זו שיטה מתקדמת לייצוג טקסטים כוקטורים (embeddings) על בסיס משמעות.
היא שונה מ-Bag of Words/TF-IDF כי היא "מבינה" דמיון סמנטי בין מילים (king ≈ queen, man ≈ woman).

```python
df_Word2Vec = df_model.copy()
```

```python
# פיצול טקסטים לרשימות מילים (Tokenization)
docs = df_Word2Vec['FullText'].astype(str).apply(lambda x: x.split())
```

 אימון מודל Word2Vec על כל הקורפוס

```python
w2v_model = Word2Vec(
    sentences=docs,
    vector_size=100,   # אפשר גם 200/300
    window=5,
    min_count=5,
    workers=4,
    sg=1,              # 1=Skip-gram, 0=CBOW
    seed=42
)
```

קבלת embedding לפוסט שלם

```python
def get_post_vector(tokens, model):
    # לוקח רק מילים שנמצאות בווקאבולרי של המודל
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

df_Word2Vec['post_vector'] = docs.apply(lambda tokens: get_post_vector(tokens, w2v_model))
```

הכנה למודלים

```python
# יצירת מטריצה מהוקטורים
post_vecs = np.vstack(df_Word2Vec['post_vector'].values)
```

פיצול Train/Test

```python
X = post_vecs
y = df_Word2Vec['buzz']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42, stratify=y)
```

* Random Forest

```python
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf))

print("=== Random Forest על Word2Vec ===")
print(classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

* XGBoost

```python
xgb = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss',
    max_depth=5,
    n_estimators=100
)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_xgb))

print("=== XGBoost על Word2Vec ===")
print(classification_report(y_test, y_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```

* Naive Bayes

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_nb))

print("=== Gaussian Naive Bayes על Word2Vec ===")
print(classification_report(y_test, y_pred_nb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_nb))
```
